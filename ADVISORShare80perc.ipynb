```python
data1 = pd.read_excel(r'India Advisor Details 2021.xlsx',sheet_name='Sheet 1')
data1
```

```python
Advisorslist = pd.DataFrame().assign(Advisor=data1['Advisor name'], AdvisorID=data1['Advisor ID'], salesvalue=data1['Sales Value'])
```

```python
Advisorslist
```

```python
DATA=Advisorslist.groupby(["Advisor", "AdvisorID"])["salesvalue"].sum().rename("Salesvalue").reset_index()
```

```python
DATA
```

```python
ADV_list=DATA.sort_values(by='Salesvalue', ascending=False)
```

```python
ADV_list
```

```python
#finding total sales
Totlsales=ADV_list["Salesvalue"].sum()
Totlsales
```

```python
TOP_Adv80=ADV_list[~ADV_list.loc[:, 'Salesvalue'].cumsum().ge(ADV_list.loc[:,'Salesvalue'].sum()*.5).shift(1, fill_value=False)].reset_index()
```

```python
TOP_Adv80.shape
```

```python
#for checking 80 percent of sales value
sumadv80=TOP_Adv80["Salesvalue"].sum()
sumadv80
```

```python
k=sumadv80/Totlsales
k
```

```python
# saving top 80% sale value advisors list in excel
file_name='Advisors2021_40Sale.xlsx'
TOP_Adv80.to_excel(file_name)
```

# Second method to find 20 percent advisors data also using CUMSUM

```python
ADV_list1=ADV_list.sort_values(by='Salesvalue', ascending=True).reset_index()
```

```python
ADV_list1
```

```python
Adv20=ADV_list1[~ADV_list1.loc[:, 'Salesvalue'].cumsum().ge(ADV_list1.loc[:,'Salesvalue'].sum()*.2).shift(1, fill_value=False)].reset_index()
```

```python
Adv20.shape
```

```python
#for checking 20 percent of sales value
sumadv20=Adv20["Salesvalue"].sum()
sumadv20
```

```python
k1=sumadv20/Totlsales
k1
```
